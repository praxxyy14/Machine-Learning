#### Logistic Regression 

Logistic Rgression is ued for Classification Only.

```python
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
import pandas as pd
df = pd.read_csv("iris.csv")
df.head()
```

```python
df["species"].unique()
```

#### Target - Species.

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
#drop the duplicate values
df = df.drop_duplicates(keep="first").reset_index(drop=True)
df.shape
```

#### Seperate X and Y (Species)

```python
X = df.drop(columns=["species"])
Y = df[["species"]]
```

```python
X.head()
```

```python
Y.head()
```

```python
Y.value_counts
```

```python
Y.value_counts().plot(kind="bar")
```

#### Apply Preprocessing on X.

```python
X.dtypes
```

##### Because only numeric features are there in X, Hence only numerical pipeline here.

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy="median"),
    StandardScaler()
).set_output(transform="pandas")
```

```python
X_pre = num_pipe.fit_transform(X)
X_pre.head()
```

##### Train,Test,Split.

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.33, random_state=21)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python

xtest.head()
```

```python
ytest.head()
```

```python
xtrain.shape
```

```python
xtest.shape
```

#### Model Building.

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(xtrain, ytrain)
```

```python

model.classes_
```

```python
model.intercept_
```

```python

model.coef_
```

#### Model evaluation

```python
# Accuracy on train
model.score(xtrain, ytrain)
```

```python
# Accuracy on test
model.score(xtest, ytest)
```

```python
ypred_test = model.predict(xtest)
ypred_test[0:5]
```

```python
ytest.head()
```

```python
yprob_test = model.predict_proba(xtest)
yprob_test[0:5]
```

```python
model.classes_
```

```python

yprob_test[0]
```

```python

from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(model, xtest, ytest)
```

```python
from sklearn.metrics import classification_report

print(classification_report(ytest, ypred_test))
```

```python

from sklearn.metrics import f1_score

f1_score(ytest, ypred_test, average="macro")
```

```python

from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, xtrain, ytrain, cv=5, scoring="f1_macro")

scores
```

```python
scores.mean()
```

##### From above metrics having good f1_macro > 0.8 hence it is good model and can be used for out of sample prediction

#### Out of sample Predicition.

```python

xnew = pd.read_csv("iris_sample.csv")
xnew
```

```python
num_pipe
```

```python
xnew_pre = num_pipe.transform(xnew)
xnew_pre
```

```python

preds = model.predict(xnew_pre)
preds
```

```python

probs = model.predict_proba(xnew_pre)
probs
```

```python
model.classes_
```

```python

xnew["species_pred"] = preds
xnew[model.classes_] = probs
```

```python

xnew = xnew.round(4)
```

```python
xnew
```

```python
xnew.to_csv("LogisticResult.csv", index=False)
```
